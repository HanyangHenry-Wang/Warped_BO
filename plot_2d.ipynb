```python
myxaxis=range(a.shape[1])
N =50

logregret_holder1_new = np.log(a)
logregret_holder2_new = np.log(b)
logregret_holder3_new = np.log(c)
logregret_holder4_new = np.log(d)
logregret_holder5_new = np.log(e)
logregret_holder6_new = np.log(f)


logregret_holder1_new[logregret_holder1_new == float('-inf')] = -8
logregret_holder2_new[logregret_holder2_new == float('-inf')] = -8
logregret_holder3_new[logregret_holder3_new == float('-inf')] = -8
logregret_holder4_new[logregret_holder4_new == float('-inf')] = -8
logregret_holder5_new[logregret_holder5_new == float('-inf')] = -8
logregret_holder6_new[logregret_holder6_new == float('-inf')] = -8
```

### 2d function 1

```python
def My_Objective_Function():

  X_total=[]

  for n in range(400*400):
    i = int(n/400)
    j = n%400
    X_total.append([i*0.015,j*0.015])


  X_total = (np. array(X_total))/10

  u = 1.2*X_total[:,0]-0.5
  v = 1.6*X_total[:,1]-0.5
  Y_total = 7.5*(1-(u**2 + v**2 - 0.3*np.cos(3*np.pi*u) - 0.3*np.cos(3*np.pi*v) ))-5.9

  Y_total = Y_total.reshape(400,400)

  return X_total, Y_total
```

```python
X_total, Y_total = My_Objective_Function()

X1 = np.linspace(0,6,400)
X2 = np.linspace(0,6,400)

X1X2, X2X1 = np.meshgrid(X1,X2)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1X2, X2X1, Y_total, cmap ='viridis', edgecolor ='green')

plt.show()
```

```python
np.max(Y_total)
```

### 2d function 2

```python
def My_Objective_Function_second():

  X_total=[]

  for n in range(40*40):
    i = int(n/40)
    j = n%40
    X_total.append([i*0.15,j*0.15])


  X_total = (np. array(X_total))

  u = X_total[:,0]-3
  v = (X_total[:,1]-3) /2
  
  Y_total = np.sin(u)*np.cos(v)*np.cos(u*v/2)+np.sin(u*v/5)

  Y_total = Y_total.reshape(40,40)

  return X_total, Y_total
```

```python
X_total, Y_total = My_Objective_Function_second()

X1 = np.linspace(0,6,40)
X2 = np.linspace(0,6,40)

X1X2, X2X1 = np.meshgrid(X1,X2)

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1X2, X2X1, Y_total, cmap ='viridis', edgecolor ='green')

plt.show()
```
